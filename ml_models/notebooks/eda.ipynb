```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option(
    "display.max_columns",
    None
)

sns.set_style(
    "whitegrid"
)

print("Libraries Loaded Successfully")
```

```python
PRODUCTS_PATH = "../datasets/products.csv"
USERS_PATH = "../datasets/users.csv"
ORDERS_PATH = "../datasets/orders.csv"
REVIEWS_PATH = "../datasets/reviews.csv"
FAKE_REVIEWS_PATH = "../datasets/fake_review.csv"
```

```python
products = pd.read_csv(
    PRODUCTS_PATH
)

users = pd.read_csv(
    USERS_PATH
)

orders = pd.read_csv(
    ORDERS_PATH
)

reviews = pd.read_csv(
    REVIEWS_PATH
)

fake_reviews = pd.read_csv(
    FAKE_REVIEWS_PATH
)

print("Datasets Loaded Successfully")
```

```python
print("Products :", products.shape)
print("Users :", users.shape)
print("Orders :", orders.shape)
print("Reviews :", reviews.shape)
print("Fake Reviews :", fake_reviews.shape)
```

```python
missing = products.isnull().sum()

missing = missing[
    missing > 0
]

plt.figure(
    figsize=(10,5)
)

missing.sort_values().plot(
    kind="barh"
)

plt.title(
    "Missing Values in Products Dataset"
)

plt.show()
```

```python
products["discounted_price"] = (
    products["discounted_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
)

products["discounted_price"] = pd.to_numeric(
    products["discounted_price"],
    errors="coerce"
)

plt.figure(figsize=(10,6))

sns.histplot(
    products["discounted_price"],
    bins=50
)

plt.title(
    "Discounted Price Distribution"
)

plt.xlabel(
    "Price (₹)"
)

plt.ylabel(
    "Number of Products"
)

plt.show()
```

```python
products["category"] \
.value_counts() \
.head(10)
```

```python
plt.figure(
    figsize=(12,6)
)

products["category"] \
.value_counts() \
.head(10) \
.plot(
    kind="bar"
)

plt.title(
    "Top Product Categories"
)

plt.show()
```

```python
products["rating"] = pd.to_numeric(
    products["rating"],
    errors="coerce"
)

plt.figure(
    figsize=(8,5)
)

sns.histplot(
    products["rating"],
    bins=20
)

plt.title(
    "Product Rating Distribution"
)

plt.xlabel(
    "Rating"
)

plt.ylabel(
    "Number of Products"
)

plt.show()
```

```python
orders.head()
```

```python
orders["order_status"] \
.value_counts()
```

```python
plt.figure(
    figsize=(8,5)
)

orders["order_status"] \
.value_counts() \
.plot(
    kind="bar"
)

plt.title(
    "Order Status Distribution"
)

plt.show()
```

```python
orders[
    "order_purchase_timestamp"
] = pd.to_datetime(
    orders[
        "order_purchase_timestamp"
    ]
)

daily_orders = (
    orders.groupby(
        orders[
            "order_purchase_timestamp"
        ].dt.date
    )
    .size()
)

plt.figure(
    figsize=(15,6)
)

daily_orders.plot()

plt.title(
    "Daily Orders Trend"
)

plt.show()
```
