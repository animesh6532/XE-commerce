```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import (
    TfidfVectorizer
)

from sklearn.metrics.pairwise import (
    cosine_similarity
)

import warnings

warnings.filterwarnings("ignore")
```

```python
products = pd.read_csv(
    "../datasets/products.csv"
)

print(
    f"Products Loaded: {len(products)}"
)
```

```python
products.head()
```

```python
products.info()
```

```python
products = products[
    [
        "product_name",
        "category",
        "about_product",
        "rating"
    ]
]

products.head()
```

```python
products.isnull().sum()
```

```python
products["about_product"] = (
    products["about_product"]
    .fillna("")
)

products["category"] = (
    products["category"]
    .fillna("")
)
```

```python
products["combined_features"] = (

    products["product_name"]

    + " "

    + products["category"]

    + " "

    + products["about_product"]

)
```

```python
products[
    [
        "product_name",
        "combined_features"
    ]
].head()
```

```python
vectorizer = TfidfVectorizer(
    stop_words="english",
    max_features=10000
)

feature_vectors = vectorizer.fit_transform(
    products["combined_features"]
)

print(
    feature_vectors.shape
)
```

```python
similarity = cosine_similarity(
    feature_vectors
)

print(
    similarity.shape
)
```

```python
product_index = pd.Series(
    products.index,
    index=products["product_name"]
)

product_index.head()
```

```python
def recommend_products(
    product_name,
    top_n=5
):

    if product_name not in product_index:

        print(
            "Product Not Found"
        )

        return

    idx = product_index[
        product_name
    ]

    similarity_scores = list(
        enumerate(
            similarity[idx]
        )
    )

    similarity_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    similarity_scores = similarity_scores[
        1 : top_n + 1
    ]

    product_indices = [

        i[0]

        for i in similarity_scores

    ]

    return products.iloc[
        product_indices
    ][
        [
            "product_name",
            "category",
            "rating"
        ]
    ]
```

```python
product_index = {}

for idx, product in enumerate(
    products["product_name"]
):
    if product not in product_index:
        product_index[product] = idx
```

```python
products[
    products["product_name"]
    .str.contains(
        "Samsung",
        case=False,
        na=False
    )
][
    [
        "product_name"
    ]
].head(10)
```

```python
recommend_products(
    "Samsung Galaxy M33 5G"
)
```

```python
products["rating"] = pd.to_numeric(
    products["rating"],
    errors="coerce"
)

products.sort_values(
    by="rating",
    ascending=False
)[
    [
        "product_name",
        "rating"
    ]
].head(10)
```
